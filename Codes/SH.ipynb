```python
import pandas as pd
import glob

# Directory containing all files
file_directory = '../Datasets/sch/*.eea'

# Initialize an empty DataFrame to store data from all files
df = pd.DataFrame()

# Channel names based on the given information
channel_names = [
    'F7', 'F3', 'F4', 'F8', 'T3', 'C3', 'Cz', 'C4', 'T4', 'T5', 'P3', 'Pz', 'P4', 'T6', 'O1', 'O2'
]

# Generate column names for all 122880 values (16 channels * 7680 samples)
column_names = [f'{channel}_Sample_{i+1}' for channel in channel_names for i in range(7680)]

# Loop through all files in the directory
for file_path in glob.glob(file_directory):
    # Read the file content
    with open(file_path, 'r') as file:
        lines = file.readlines()
    
    # Convert lines to float values
    numeric_data = [float(line.strip()) for line in lines if line.strip()]
    
    # Extract the file name without extension
    file_name = file_path.split('/')[-1].split('.')[0]
    
    # Extract the label from the file name 
    label= 'Schizophrenia'
    data_row = pd.DataFrame([[label] + numeric_data], index=[file_name], columns=['Label'] + column_names)
    # Append to the main DataFrame
    df = pd.concat([df, data_row, ])
    
file_directory = '../Datasets/norm/*.eea'
# Loop through all files in the directory
for file_path in glob.glob(file_directory):
    # Read the file content
    with open(file_path, 'r') as file:
        lines = file.readlines()
    
    # Convert lines to float values
    numeric_data = [float(line.strip()) for line in lines if line.strip()]
    
    # Extract the file name without extension
    file_name = file_path.split('/')[-1].split('.')[0]
    
    # Reshape data to have a single row of 122880 values (16 channels * 7680 samples)
    label= 'HC'
    data_row = pd.DataFrame([[label] + numeric_data], index=[file_name], columns=['Label'] + column_names)
    
    # Append to the main DataFrame
    df = pd.concat([df, data_row])

df
```

```python
# Plot example EEG signals from a healthy subject and a subject with schizophrenia
healthy_subject = df[df['Label'] == 'HC'].iloc[0]
schiz_subject = df[df['Label'] == 'Schizophrenia'].iloc[0]

plt.figure(figsize=(12, 6))
plt.plot(range(7680), healthy_subject[1:7681], label='Healthy (F7)', alpha=0.7)
plt.plot(range(7680), schiz_subject[1:7681], label='Schizophrenia (F7)', alpha=0.7)
plt.title('EEG Signals from F7 Channel (First Minute)')
plt.xlabel('Sample Number')
plt.ylabel('EEG Amplitude (mkV)')
plt.legend()
plt.show()
```

```python
f_healthy, psd_healthy = welch(healthy_subject[1:7681], fs=128)
f_schiz, psd_schiz = welch(schiz_subject[1:7681], fs=128)

# Ensure the PSD values are real by taking the absolute value
psd_healthy = np.abs(psd_healthy)
psd_schiz = np.abs(psd_schiz)

plt.figure(figsize=(10, 5))
plt.semilogy(f_healthy, psd_healthy, label='Healthy (F7)', alpha=0.7)
plt.semilogy(f_schiz, psd_schiz, label='Schizophrenia (F7)', alpha=0.7)
plt.title('Power Spectral Density (PSD) of F7 Channel')
plt.xlabel('Frequency (Hz)')
plt.ylabel('PSD (V^2/Hz)')
plt.legend()
plt.show()
```

```python
def extract_features(row):
    features = {}
    for channel in channel_names:
        channel_data = row[[col for col in row.index if channel in col]]
        features[f'{channel}_mean'] = channel_data.mean()
    return pd.Series(features)

# Apply feature extraction to all subjects
features_df = df.apply(extract_features, axis=1)
features_df['Label'] = df['Label']
```

```python
from sklearn.metrics import accuracy_score, recall_score, confusion_matrix, f1_score, precision_score

def evaluate_model(model, x, y_true):
    y_pred = model.predict(x)
    return {
        'Accuracy': accuracy_score(y_true, y_pred),
        'Precision': precision_score(y_true, y_pred),
        'Recall': recall_score(y_true, y_pred),
        'F1 Score': f1_score(y_true, y_pred),
        'Confusion Matrix': confusion_matrix(y_true, y_pred)
    }
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_confusion_matrices(matrices, titles):
    fig, axes = plt.subplots(1, len(matrices), figsize=(6 * len(matrices), 4))
    for i, (cm, title) in enumerate(zip(matrices, titles)):
        ax = axes[i] if len(matrices) > 1 else axes
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax)
        ax.set_title(title)
        ax.set_ylabel('Actual')
        ax.set_xlabel('Predicted')
        print(title,":\n",cm)
    plt.tight_layout()
    plt.show()
```
